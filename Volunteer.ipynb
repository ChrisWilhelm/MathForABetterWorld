```python
import pandas as pd
from datetime import datetime, timedelta
from json import loads, dumps

# # FOR VOLUNTEER DATA
df = pd.read_csv('2024Volunteer.csv')
times = []
startTimes = []
i = 0
for i, row in df.iterrows():
    dateString = row['DATE']
    datetime_object = datetime.strptime(dateString, '%m.%d.%Y %H:%M:%S')
    startTimes.append(datetime.strftime(datetime_object, '%Y-%m-%d %H:%M:%S'))
    datetime_object = datetime_object + timedelta(hours=4)
    times.append(datetime.strftime(datetime_object, '%Y-%m-%d %H:%M:%S'))

df['start'] = startTimes
df['end'] = times

df = df.rename(columns={'VOLUNTEER': 'name', 'WEIGHT': 'regularTaken'})

df = df[['start', 'end', 'name', 'regularTaken']]

# print("\n".join(df["name"].astype(str).unique()))
# print("\n".join(df["name"].unique()))
print('[')
for i, row in df.iterrows():
    print('\t{')
    print('\t\tstart: ' + '"' + row['start'] + '"' + ',')
    print('\t\tend: ' + '"'+ row['end'] + '"' +',')
    print('\t\tname: ' + '"'+ row['name'] + '"' +',')
    print('\t\tregularTaken: ' + str(row['regularTaken']) + ',',)
    print('\t\tdamagedTaken: 0,')
    print('\t},')
print(']')
```
